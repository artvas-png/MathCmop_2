```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
N = 500 #x axis
M = 500 #y axis
h = 1 / max(N, M)
Tend = 0.0005
Nt = 1000
dt = Tend/Nt
u= np.zeros((Nt, N, M))
g1 = np.zeros(M) + 0 #u[0, y]
g2 = np.zeros(M) + 0 #u[1, y] 
g3 = np.zeros(N) + 0 #u[x, 0]
g4 = np.zeros(N) + 0 #u[x, 1]
u[0] = g3
u[N-1] = g4
u[:, 0] = g1
u[:, M-1] = g2
```

```python
def f(x, y):
    A = abs(x - 0.5)+ abs(y - 0.5) < 0.05
    return 50 *  np.array(A)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, M, endpoint=False), np.linspace(0, 1, N, endpoint=False))
plt.imshow((f(X, Y)), cmap='Greens')
```

```python
for i in range(1, Nt):
    u[i, 1:N-1, 1:M-1] = u[i-1, 1:N-1, 1:M-1] + dt * (f(X[1:N-1, 1:M-1], Y[1:N-1, 1:M-1]) + (u[i-1, 2:N, 1:M-1] 
                                + u[i-1, 0:N-2, 1:M-1] + u[i-1, 1:N-1, 2:M]
                                                    + u[i-1, 1:N-1, 0:M-2] - 4 * u[i-1, 1:N-1, 1:M-1]) / h**2)
```
